```python
from pandas import DataFrame


def parser(line, trig = False ):
    info = []
    if(trig):
        info = line.split(',')[:10]
        for e,i in enumerate(info):
            if e == 0:
                info[e] = float(i[2:7])
            else:    
                info[e] = float(i[3:8]) 
    else:
        info = line.split(',')[:10]
        for e,i in enumerate(info):
            info[e] = float(i)            
    return info            
    

file_name = 'hw1/speed_O{0}_{1}.txt'
hosts = ['trig', '336', 'home']
all_data ={}
for host in hosts:
    for i in range(4):
        my_file = file_name.format(str(i),host)
        f = open(my_file)
        trig = False
        if host == 'trig':
            trig = True
        data = {}
        previous_line =''
        for line in f:
            #print(line)
            if(previous_line == 'system\n'):
                data['system'] = parser(line, trig)
            if(previous_line == 'user\n'):       
                data['user'] = parser(line, trig)
            if(previous_line == 'elapsed\n'):     
                data['elapsed'] = parser(line, trig)
            previous_line = line    

        all_data[my_file[4:]] = data
all_data
```
