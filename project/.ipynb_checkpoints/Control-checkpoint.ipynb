```python
def get_num_clusters(alpha):
    my_string = "./look_up {0}".format(alpha)
    x = write(my_string)
    #print(x)
    z = x.split("\n")[2]
    if(z[:15] == 'num_of_clusters'):
        return int(z[15:])
    else:
        return np.infty

get_num_clusters(0.89)
```

```python
x = linspace(0.5, 1.5,20)
clusters = [get_num_clusters(i) for i in x]
clusters
```

```python
figsize(15,7)
x = linspace(0.8, 1.0,40)
clusters = [get_num_clusters(i) for i in x]
plot(x,clusters)
```
